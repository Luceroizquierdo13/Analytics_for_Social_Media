```python
#!/usr/bin/env python
# coding: utf-8
import urllib.request
import re
import schedule
import time
from datetime import datetime
import numpy as np

# Category:  
print("RETAIL: COSMETIC PRODUCTS ")
print("PRODUCT: Loreal true match le blush")
print("BRAND: Loreal Paris")
print("START:",str(datetime.now()),"\n")
print("Loreal true match le blush prices in US\n")

#****************************** PRODUCT 4: Loreal true match le blush ***************************************************
def jobblushprice():
    currenttime = str(datetime.now())
    print("(",currenttime[0:19],")")
    # 1. COMPANY: WELL.CA COMPANY
    htmlfile = urllib.request.urlopen("https://well.ca/products/loreal-paris-true-match_4395.html")
    htmltext = htmlfile.read()
    WellCA_blush = re.findall(r'<div class = "currentPrice noSalePrice">(.*?)</div>',str(htmltext))
    print("Best price in WELL CA :", WellCA_blush[0], "; Overall Rating : N/A")

    # 2. COMPANY: ULTA BEAUTY
    htmlfile = urllib.request.urlopen("https://www.ulta.com/p/true-match-super-blendable-blush-VP10730")
    htmltext = htmlfile.read()
    Beauty_blush = re.findall(r'<span class="Text Text--title-6 Text--left Text--bold Text--small Text--neutral-80"><label class="sr-only">Sale Price</label>(.*?)</span>',str(htmltext))
    #because the rating in the source webpage is not visible, I am typing this one
    print("Best price in ULTA BEAUTY :", Beauty_blush,"; Overall Rating : 4.11 / 5 stars")

    # 3. COMPANY: CARE TO BEAUTY
    htmlfile = urllib.request.urlopen("https://www.caretobeauty.com/us/l-oreal-paris-accord-parfait-le-blush-150-candy-cane-pink-blush/")
    htmltext = htmlfile.read()
    Carebeauty_blush = re.findall(r'<span class="price">(.*?)</span>',str(htmltext))
    Carebeauty_rating = re.findall(r'<div class="information-bar__content"><h6 class=""(.*?)</h6>',str(htmltext))
    Carebeauty_rating = [elem.replace('>', '') for elem in Carebeauty_rating]
    print("Best price in CARE TO BEAUTY :", Carebeauty_blush[0],"; Overall Rating :",Carebeauty_rating[2])

     # 4. COMPANY: LADY MAKEUP
    htmlfile = urllib.request.urlopen("https://www.ladymakeup.com/shop/l_oreal_le_blush_true_match.html")
    htmltext = htmlfile.read()
    Lady_blush = re.findall(r'<span class="price nowrap">(.*?)</span>',str(htmltext))
    Lady_blush = [elem.replace('>', '') for elem in Lady_blush]
    print("Best price in LADY MAKEUP :", Lady_blush[0],"; Overall Rating : 4.88/5")
    print("\n")

schedule.every().day.at("20:05").do(jobblushprice)

while 1:
    schedule.run_pending()
    time.sleep(1)
```
