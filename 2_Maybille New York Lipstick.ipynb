```python
#!/usr/bin/env python
# coding: utf-8
import urllib.request
import re
import schedule
import time
from datetime import datetime
import numpy as np

print("RETAIL: COSMETIC PRODUCTS ")
print("PRODUCT: Maybelline color sensational ultimatte slim lipstick ")
print("BRAND: Maybelline New York")
print("START: ",str(datetime.now()),"\n")
print("Maybelline color sensational ultimatte slim lipstick prices in US\n")

#****************************** PRODUCT 2: maybelline color sensational ultimatte slim lipstick ***************************************************
def joblipstickprice():
    currenttime = str(datetime.now())
    print("(",currenttime[0:19],")")
    # 1. COMPANY: ULTA BEAUTY
    htmlfile = urllib.request.urlopen("https://www.ulta.com/p/color-sensational-ultimatte-slim-lipstick-pimprod2020630")
    htmltext = htmlfile.read()
    Beauty_lipstick = re.findall(r'<span class="Text Text--title-6 Text--left Text--bold Text--small Text--neutral-80"><label class="sr-only">Price</label>(.*?)</span>',str(htmltext))
    #because the rating in the source webpage is not visible, I am typing this one
    print("Best price in ULTA BEAUTY : ", Beauty_lipstick[0],"; Overall Rating : 4.47 / 5 stars")

    # 2. COMPANY: LOOK FANTASTIC COMPANY
    htmlfile = urllib.request.urlopen("https://us.lookfantastic.com/maybelline-colour-sensational-ultimatte-slim-lipstick-25g-various-shades/13496155.html")
    htmltext = htmlfile.read()
    Look_lipstick = re.findall(r'<span class="productPrice_schema productPrice_priceAmount" data-schema="price">(.*?)</span>',str(htmltext))
    LOOK_rating = re.findall(r'<span class="athenaProductReviews_aggregateRating_hiddenLabel">(.*?)</span>',str(htmltext))
    print("Best price in LOOKFANTASTIC : $",Look_lipstick[0],"; Overall Rating : N/A" )

    # 3. COMPANY: CARE TO BEAUTY 
    htmlfile = urllib.request.urlopen("https://www.caretobeauty.com/us/maybelline-color-sensational-lipstick-144-naked-dare")
    htmltext = htmlfile.read()
    Carebeauty_lipstick = re.findall(r'<span class="price">(.*?)</span>',str(htmltext))
    Carebeauty_rating = re.findall(r'<div class="information-bar__content"><h6 class=""(.*?)</h6>',str(htmltext))
    Carebeauty_rating = [elem.replace('>', '') for elem in Carebeauty_rating]
    print("Best price in CARE TO BEAUTY : $", Carebeauty_lipstick[0], "; Overall Rating :",Carebeauty_rating[2])

    # 4. COMPANY: WELLCA
    htmlfile = urllib.request.urlopen("https://well.ca/products/maybelline-color-sensational_199496.html")
    htmltext = htmlfile.read()
    WellCA_lipstick = re.findall(r'<div class = "currentPrice noSalePrice">(.*?)</div>',str(htmltext))
    print("Best price in WELLCA :", WellCA_lipstick[0],"; Overall Rating : N/A"  )

    # 5. COMPANY: Ebay
    htmlfile = urllib.request.urlopen("https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2334524.m570.l1313&_nkw=Maybelline+Color+Sensational+Ultimatte+Slim+Lipstick&_sacat=0&LH_TitleDesc=0&_odkw=Maybelline+Color+Sensational+Made+For+All+Lipstick&_osacat=0&LH_PrefLoc=3")
    htmltext = htmlfile.read()
    Ebay_lipstick = re.findall(r'<span class=s-item__price>(.*?)</span>',str(htmltext))
    Ebay_lipstick = [elem.replace('<span class=DEFAULT> to ', '') for elem in Ebay_lipstick]
    Ebay_lipstick = [elem.replace('<span class=ITALIC>', '') for elem in Ebay_lipstick]

    Ebay_lipstick = [elem.replace('$', '') for elem in Ebay_lipstick]
    Ebay_lipstick=  [float(elem) for elem in Ebay_lipstick]
    Ebay_lipstick.sort()

    # Rating
    Ebay_rating = re.findall(r'<span class=clipped>(.*?)</span',str(htmltext))
    random_list =[]
    frequency = {}
    for x in Ebay_rating: 
        if 'out of' in x:
            random_list.append(x)

    # iterating over the list
    for item in random_list:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1
    index_min = np.argmin(Ebay_lipstick)
    print("Best price in Ebay : $", Ebay_lipstick[index_min],"; Overall Rating :",frequency)
    print("\n")

schedule.every().day.at("20:05").do(joblipstickprice)

while 1:
    schedule.run_pending()
    time.sleep(1)
```
